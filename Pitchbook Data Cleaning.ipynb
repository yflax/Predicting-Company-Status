```python
def binarize(df, column):
    # Iterate through each value in the specified column
    for i, value in enumerate(df[column]):
        #i represents the index of the current row being iterated over, and 'value' represents the value of the element in the specified column for that particular row.
        # Check if the value is not equal to 'Profitable'
        if value != 'Profitable':
            # If not 'Profitable', impute 'Not Profitable'
            df.at[i, column] = 'Not Profitable'
    
    # Return the modified DataFrame
    return df[column]
```
